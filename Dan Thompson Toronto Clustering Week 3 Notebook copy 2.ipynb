```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis 
import codecs
from bs4 import BeautifulSoup

!wget https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M
        
```

#Now parse the neighborhood table into a string, and turn that into a DataFrame, name the columns and drop unassigned

```python
f = codecs.open('List_of_postal_codes_of_Canada:_M', encoding='utf-8')
html = f.read()
soup = BeautifulSoup(html, 'html.parser')

df = (pd.read_html(str(soup.table))[0])

#drop the first row, which is the headings of the table
df.drop(axis=0, index=0, inplace=True)

#then assign the proper column names
column_names = ['Postcode', 'Borough', 'Neighborhood'] 
df.columns = column_names

#get rid of unassigned Boroughs
df = df[df['Borough'] != "Not assigned"].reset_index(drop=True) 
df
```

#If the neighborhood isn't assigned, make it the same as the borough name 

```python
na = df[df['Neighborhood']=="Not assigned"]
for index in range(len(na)):
    r = na.iloc[index]
    r['Neighborhood']=r['Borough']
    df.iloc[na.index[0]]['Neighborhood'] = r['Borough']

df.head(8)    
```

```python
#Now get all of the unique Post codes using Groupby, then walk through that df. 
#Match the post code in the original df
#Set the Borough, and then make a list of all the neighborhoods in that postcode, format the string using join method
```

```python
g = df.groupby('Postcode').count().reset_index() 
for index in range(len(g)):
    r = g.iloc[index]
    p = r.Postcode
    d = (df[df['Postcode'] == p])
    r.Borough = d.iloc[0]['Borough']
    r.Neighborhood = ', '.join(list(d['Neighborhood']))
    g.iloc[index] = r

#this throws a warning but the results seem to check out    
```
